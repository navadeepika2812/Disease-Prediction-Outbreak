Importing Modules

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import pickle
```

```python
diabetes_dataset=pd.read_csv("C:/Users/HP/Desktop/PDO1/Dataset/diabetes.csv")
```

```python
x=diabetes_dataset.drop("Outcome",axis=1)
y=diabetes_dataset['Outcome']
```

```python
print(x)
```

```python
print(y)
```

Data Standardization

```python
scaler=StandardScaler()
```

```python
scaler.fit(x)
```

```python
std=scaler.transform(x)
```

```python
X=std
Y=diabetes_dataset['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

Train_Test_Split

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Training the Model

```python
model=svm.SVC(kernel="linear")
```

```python
model.fit(X_train,Y_train)
```

Model Evalution

Accuracy

```python
X_train_predict=model.predict(X_train)

```

```python
accuracy_score(X_train_predict,Y_train)
```

```python
X_test_predict=model.predict(X_test)
```

```python
accuracy_score(X_test_predict,Y_test)
```

```python
classification_report(X_test_predict,Y_test)
```

```python
cm=confusion_matrix(X_test_predict,Y_test)
```

```python
print(cm)
```

```python
cmd = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Class 1", "Class 2"])
cmd.plot(cmap= 'BuGn')

plt.title("Confusion Matrix for Diabetes")

plt.show()
```

```python

feature_names=["Pregnancies","Glucose","BloodPressure","SkinThickness","Insulin","BMI","DiabetesPedigreeFunction","Age"]
input_data=[8,133,72,0,0,32.9,0.27,39]
input_data_as_df = pd.DataFrame([input_data], columns=feature_names)


input_data_as_numpy_array=np.asarray(input_data)
input_data_reshape=input_data_as_numpy_array.reshape(1,-1)
std_data=scaler.transform(input_data_reshape)
predict=model.predict(std_data)


if (predict[0] == 1):
    print("The Person has diabetes disease")
else:
    print("The Person does not  have diabetes disease")
```
